```python
# Now we read data in JSON format and display the histogram of users per processes.
def get_jsonparsed_data(url):
    response = urlopen(url)
    data = response.read().decode("utf-8")
    return json.loads(data)

histo = dict()

# We know the data available in this JSON document.
url = ("http://vps516494.ovh.net/Survol/survol/sources_types/enumerate_CIM_Process.py?xid=.&mode=json")
lstProc = get_jsonparsed_data(url)
for oneNod in lstproc['nodes']:
    if oneNod['entity_class'] == 'CIM_Process':
        try:
            usr = oneNod['survol_info_dict']['user']
        except KeyError:
            pass
        try:
            histo[usr] += 1
        except KeyError:
            histo[usr] = 1

plt.bar(range(len(histo)), histo.values(), align='center')
plt.xticks(range(len(histo)), histo.keys(), rotation=25)

plt.show()
```
