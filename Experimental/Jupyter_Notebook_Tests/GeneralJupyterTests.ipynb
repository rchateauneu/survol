```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
t = np.arange(0,2*np.pi,0.1)   # start,stop,step
x = np.sin(t)
y = np.cos(t)
z = np.sin(2*t)
plt.plot(t,x,t,y,t,z)
plt.show()
```

```python
t = np.arange(0,2*np.pi,0.1)   # start,stop,step
vec = []
for i in range(20):
    x = np.sin(t*(i+1))
    vec.append(t)
    vec.append(x)
plt.plot(*vec)
plt.show()
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.title("Courbe de Lissajous")
t = np.arange(0,2*np.pi,0.01)   # start,stop,step
px = 19
py = 17
x = np.sin(t*px)
y = np.cos(t*py)
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.title("Courbe du Scarabee")
t = np.arange(0,2*np.pi,0.01)   # start,stop,step
a = 5
l = 5*a/3
r = l * np.cos(2*t)- a * np.cos(t)
x = r * np.cos(t)
y = r * np.sin(t)
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.title("Courbe du Diable")
t = np.arange(0,2*np.pi,0.001)   # start,stop,step
a = 50
g = 4
r = np.sqrt( 4*g*a*a + a*a / np.cos(2*t) )
x = r * np.cos(t)
y = r * np.sin(t)
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.title("r = cos(7 theta) + 3")
theta = np.arange(0,2*np.pi,0.001)   # start,stop,step
r = np.cos( 7 * theta ) + 3
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.title("r = cos( 3 * theta / 10.0) + 0.2")
theta = np.arange(0,20*np.pi,0.0001)   # start,stop,step
r = np.cos( 3 * theta / 10.0) + 0.2
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.plot(x,y)
plt.show()
```

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import time

def pltsin(ax, px, py, colors=['b']):
    x = np.sin(t*px)
    y = np.cos(t*py)

    if ax.lines:
        for line in ax.lines:
            line.set_xdata(x)
            line.set_ydata(y)
    else:
        for color in colors:
            ax.plot(x, y, color)
    fig.canvas.draw()

t = np.arange(0,2*np.pi,0.01)   # start,stop,step

fig,ax = plt.subplots(1,1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
for px in range(5,23):
    for py in range(5,15):
        pltsin(ax, px, py, ['b', 'r'])
        time.sleep(0.01)
```
